### Бешенство (rabies)

Прирост количества патогенов за счет размножения и их убыль за счет нейтрализации антителами:
$\\$

$\\$
$\mathcal{\frac{dV}{dt} = \beta V - \gamma FV}$ (1) 
$\\$

$\\$
Изменение количества плазматических клеток:
$\\$

$\\$
$\mathcal{\frac{dC}{dt} = \alpha (1 - m(t)) F(t - \tau)V(t - \tau) - \mu(C - C*)}$ (2)
$\\$

$\\$
Изменение количества антител:
$\\$

$\\$
$\mathcal{\frac{dF}{dt} = \rho C - \eta \gamma F V - \mu_f F}$ (3)
$\\$

$\\$
Изменение степени поражения органа-мишени (ЦНС):
$\\$

$\\$
$\mathcal{\frac{dm}{dt} = \sigma V - \mu_m m}$ (4)
$\\$

$\\$
$\mathcal{V(t)}$  - количество размножающихся патогенов в момент t
$\\$
$\mathcal{F(t)}$  - количество антител (иммуноглобулины, иммуноглобулиновые рецепторы...) в момент t
$\\$
$\mathcal{C(t)}$  - количество плазматических клеток (носителей и продуцентов антител) в момент t
$\\$
$\mathcal{m(t) = 1 - \frac {M(t)} {M*}}$  - доля поражения органа-мишени (M(t) - количество клеток в момент t, M* - количество клеток в норме)
$\\$
$\mathcal{\tau}$ - время деления плазматических клеток
$\\$
$\mathcal{\alpha (1 - m(t)) F(t - \tau)V(t - \tau)}$ - образование плазматических клеток
$\\$
$\mathcal{\mu(C - C*)}$ - гомеостаз плазматических клеток за счет притока из костного мозга и естественного старения
$\\$
$\mathcal{\rho C}$ - продукция антител плазматическими клетками
$\\$
$\mathcal{\eta \gamma F V}$ - расход антител на нейтрализацию патогенов
$\\$
$\mathcal{\mu_f F}$ - изменение количества антител за счет разрушения
$\\$
$\mathcal{\sigma V}$ - поражение органа-мишени (ЦНС) из-за патогена
$\\$
$\mathcal{\mu_m m}$ - уменьшение степени поражения органа-мишени (ЦНС) из-за регенерации
$\\$

Прирост количества патогенов за счет размножения и их убыль за счет нейтрализации антителами:
$\\$

$\\$
$\mathcal{\beta V >> \gamma FV}$ (1) 
$\\$

$\\$
Поражение органа-мишени (ЦНС) при проявлении клинических признаков:
$\\$

$\\$
$\mathcal{\sigma V = 100\%}$ (4)
$\\$
